## fun-asr-nano-2512 模型测试记录

```python
from funasr import AutoModel

model_dir = "FunAudioLLM/Fun-ASR-Nano-2512"

model = AutoModel(
    model=model_dir,
    # vad_model="fsmn-vad",
    # vad_kwargs={"max_single_segment_time": 30000},
    device="cuda:0",
    disable_update=True,
    trust_remote_code=True,
    remote_code="./fun_asr_code/model.py",
)
```

```python
wav_path = "D:\\Music\\asr_test\\test_asr.mp3"
res = model.generate(input=[wav_path], cache={}, batch_size_s=10)
print(res)
```

```python
for key, value in res[0].items():
    print(f"{key}: {value}")
    if key == "timestamps":
        for timestamp in value:
            # Timestamp: {'token': 'V', 'start_time': 0.06, 'end_time': 0.12, 'score': 0.0}
            # Timestamp: {'token': 'al', 'start_time': 0.3, 'end_time': 0.36, 'score': 0.068}
            print(f"\tTimestamp: {timestamp}")
```
